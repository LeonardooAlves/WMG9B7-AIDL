## Verify Setup

Run the setup verification notebook (`1 - Setup.ipynb`) to ensure your API key is properly configured.

````python
# Quick test to verify your Groq API key works
import os
from dotenv import load_dotenv

load_dotenv("../.env")

if os.getenv("GROQ_API_KEY"):
    print("✅ GROQ_API_KEY found in environment")
    
    try:
        from openai import OpenAI
        client = OpenAI(
            base_url="https://api.groq.com/openai/v1",
            api_key=os.getenv("GROQ_API_KEY")
        )
        
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[{"role": "user", "content": "Say hello in one sentence."}],
            max_tokens=20
        )
        
        print("✅ Groq API connection successful!")
        print(f"Response: {response.choices[0].message.content}")
        
    except Exception as e:
        print(f"❌ API connection failed: {e}")
else:
    print("❌ GROQ_API_KEY not found. Please add it to your .env file.")
````
